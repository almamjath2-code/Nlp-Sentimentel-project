This project is a simple Natural Language Processing (NLP) based sentiment analysis system. The main goal of this project is to classify input text into three categories: Positive, Negative, and Neutral. The model is built using machine learning techniques with TF-IDF vectorization for feature extraction and a trained classification model for prediction.

The workflow starts with taking raw text input from the user. The text is first cleaned using basic preprocessing techniques such as lowercasing, removing unwanted characters, and normalizing the text. After that, lemmatization is applied using POS tagging to reduce words to their root form, which helps improve model accuracy by reducing noise in the data.

Once the text is preprocessed, it is converted into numerical form using a TF-IDF vectorizer that transforms the text into meaningful features. These features are then passed into a trained machine learning model, which predicts the sentiment of the given text.

Finally, the output is displayed as either Positive, Negative, or Neutral based on the model’s prediction. This project demonstrates a complete NLP pipeline including data cleaning, feature engineering, and machine learning prediction in a simple and practical way.
